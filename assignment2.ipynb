## 3.1 Splitting the data into a training set and a test set. Specifically, you need to split the data at the following ratios, respectively, to form three different suites of training and test sets:

### i) Converting DataFrame to 2D-Numpy Array 'X' and Target Numpy Array 'y':

```python
patient_record=pd.read_csv('heart_failure_clinical_records_dataset.csv')

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

#converting data to numpy and storing all feature vectors in X
df1=patient_record[['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes','ejection_fraction', 
                    'high_blood_pressure', 'platelets','serum_creatinine','serum_sodium', 'sex', 'smoking']]
X=df1.to_numpy()
#converting DEATH_EVENT (target column) to numpy array y
arr=patient_record.to_numpy()
y=arr[:,12]

print(X.shape)
print(y.shape)
```

### ii) Creating Test Suites From the Data

```python
X=MinMaxScaler().fit(X).transform(X) #scaling the data for KNN as it is a distance based algorithm

X_train1, X_test1, y_train1, y_test1=train_test_split(X, y, test_size=0.5, random_state=1)
print(f'Suite 1: Training {X_train1.shape}, Testing {X_test1.shape}')

X_train2, X_test2, y_train2, y_test2=train_test_split(X, y, test_size=0.4, random_state=1)
print(f'Suite 2: Training {X_train2.shape}, Testing {X_test2.shape}')

X_train3, X_test3, y_train3, y_test3=train_test_split(X, y, test_size=0.2, random_state=1)
print(f'Suite 3: Training {X_train3.shape}, Testing {X_test3.shape}')
```

### i) Model 1: K-Nearest Neighbour Classifier Hyperparameter Selection:



```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.model_selection import cross_val_score  


def tune_hyper_parameters (X_train,y_train):
    neighbors = list(range(3, 9,2))
    cv_scores =[]

    for k in neighbors:
        knn =KNeighborsClassifier(n_neighbors=k)
        scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='f1_macro')
        print(f'Accuracy for n_neighbors={k} : {scores.mean()}')
        cv_scores.append(scores.mean())
    
    k_best = neighbors[np.argmax(cv_scores)]  
    print(f"\n\n Best cv_score is with {k_best} neighbours: {np.max(cv_scores)}")

    plt.figure(figsize=(7,5))
    plt.xticks(range(3,9, 2))
    plt.title('Optimizing value of hyperparameter n_neighbors', fontsize=15)
    sns.lineplot(data=cv_scores, y=cv_scores, x=neighbors, marker='o')
    plt.savefig('27_TuningKNN')
    plt.show()
    return k_best

n1=tune_hyper_parameters(X_train1, y_train1)
n2=tune_hyper_parameters(X_train2, y_train2)
n3=tune_hyper_parameters(X_train3, y_train3)


```

### Model fitting for KNN and Performance Metrics:

```python

'''
CONSIDERING ALL THREE TEST SUITS 
n-neighbours=11 is considered the most optimal value
as it does not lead to unexpected F1 and Precision Score values
due to labels with no predicted samples leading to division by zero.
'''

knn3=KNeighborsClassifier(n_neighbors=n1)
print('='*40+' SUITE 1 (50:50) KNN CLASSIFIER ' +'='*40)
model1=knn3.fit(X_train1, y_train1)
y_pred1=model1.predict(X_test1)
cm1=confusion_matrix(y_test1, y_pred1)
print(f'Confusion Matrix: \n{cm1}\n\n\n')
ac1=accuracy_score(y_test1, y_pred1)
print(f'Accuracy Score: {ac1}\n\n\n')
cr=classification_report(y_test1, y_pred1)
print(f'Classification Report: \n\n{cr}')
crd_knn_1=classification_report(y_test1, y_pred1, output_dict=True)



knn3=KNeighborsClassifier(n_neighbors=n2)
print('='*40+' SUITE 2 (60:40) KNN CLASSIFIER ' +'='*40)
model2=knn3.fit(X_train2, y_train2)
y_pred2=model2.predict(X_test2)
cm2=confusion_matrix(y_test2, y_pred2)
print(f'Confusion Matrix: \n{cm2}\n\n\n')
ac2=accuracy_score(y_test2, y_pred2)
print(f'Accuracy Score: {ac2}\n\n\n')
cr=classification_report(y_test2, y_pred2)
print(f'Classification Report: \n\n {cr}')
crd_knn_2=classification_report(y_test2, y_pred2, output_dict=True)


knn3=KNeighborsClassifier(n_neighbors=n3)
print('='*40+' SUITE 3 (80:20) KNN CLASSIFIER ' +'='*40)
model3=knn3.fit(X_train3, y_train3)
y_pred3=model3.predict(X_test3)
cm3=confusion_matrix(y_test3, y_pred3)
print(f'Confusion Matrix: \n{cm3}\n\n\n')
ac3=accuracy_score(y_test3, y_pred3)
print(f'Accuracy Score: {ac3}\n\n\n')
cr=classification_report(y_test3, y_pred3)
print(f'Classification Report: \n\n{cr}\n\n')
crd_knn_3=classification_report(y_test3, y_pred3, output_dict=True)
```

### ii) Model 2: Decision Tree Classifier Parameter Selection (max_depth):



```python
patient_record=pd.read_csv('heart_failure_clinical_records_dataset.csv')

from sklearn import preprocessing
from sklearn.model_selection import train_test_split
df1=patient_record[['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes','ejection_fraction', 
                    'high_blood_pressure', 'platelets','serum_creatinine','serum_sodium', 'sex', 'smoking']]
X=df1.to_numpy()
arr=patient_record.to_numpy()
y=arr[:,12]
X_train1, X_test1, y_train1, y_test1=train_test_split(X, y, test_size=0.5, random_state=1)
X_train2, X_test2, y_train2, y_test2=train_test_split(X, y, test_size=0.4, random_state=1)
X_train3, X_test3, y_train3, y_test3=train_test_split(X, y, test_size=0.2, random_state=1)



from sklearn.tree import DecisionTreeClassifier
#TUNING MAX_DEPTH
def tune_d_tree(X_train, y_train):
    
    accuracy_scores=[]
    best_depth=0
    best_cv_score=0
    depth_values=list(range(2,20))

    for depth in depth_values:
        d_tree=DecisionTreeClassifier(max_depth=depth)
        cv_scores=cross_val_score(d_tree,X_train, y_train, cv=5, scoring='accuracy')
        print(f'Depth: {depth} cv_scores:{cv_scores.mean()}')
        if best_cv_score<cv_scores.mean():
            best_cv_score=cv_scores.mean()
            best_depth=depth
        accuracy_scores.append(cv_scores.mean())
    print(f'Best Depth= {best_depth}')
    print(f'Best Accuracy Score: {best_cv_score}')
    #print(f'All scores: \n{accuracy_scores}')
    plt.figure(figsize=(15,5))
    plt.xticks(range(2,20))
    plt.title('Optimizing value of hyperparameter max_depth', fontsize=15)
    sns.lineplot(data=accuracy_scores, y=accuracy_scores, x=depth_values, marker='o')
    plt.savefig('28_tuning_max_depth')
    plt.show()
    print('\n')
    return best_depth

print('='*40 + 'TEST SUITE 1' + '='*40)    
max_d1=tune_d_tree(X_train1, y_train1)
print('='*40 + 'TEST SUITE 2' + '='*40)    
max_d2=tune_d_tree(X_train2, y_train2)
print('='*40 + 'TEST SUITE 3' + '='*40)    
max_d3=tune_d_tree(X_train3, y_train3)
```

```python
#TUNING MIN_SAMPLES_LEAF
def tune_d_tree2(X_train, y_train):
    
    accuracy_scores=[]
    best_msl=0
    best_cv_score=0
    msl=list(range(1,15))

    for leaves in msl:
        d_tree=DecisionTreeClassifier(min_samples_leaf=leaves)
        cv_scores=cross_val_score(d_tree,X_train, y_train, cv=5, scoring='accuracy')
        print(f'Depth: {leaves} cv_scores:{cv_scores.mean()}')
        if best_cv_score<cv_scores.mean():
            best_cv_score=cv_scores.mean()
            best_msl=leaves
        accuracy_scores.append(cv_scores.mean())
    print(f'Best Min. Sample Leaf= {best_msl}')
    print(f'Best Accuracy Score: {best_cv_score}')
    #print(f'All scores: \n{accuracy_scores}')
    plt.figure(figsize=(15,5))
    plt.xticks(range(1,15))
    plt.title('Optimizing value of hyperparameter min_samples_leaf', fontsize=15)
    sns.lineplot(data=accuracy_scores, y=accuracy_scores, x=msl, marker='o')
    plt.savefig('29_tuning_min_samples')
    plt.show()
    print('\n')
    return best_msl

print('='*40 + 'TEST SUITE 1' + '='*40)    
leaf1=tune_d_tree2(X_train1, y_train1)
print('='*40 + 'TEST SUITE 2' + '='*40)    
leaf2=tune_d_tree2(X_train2, y_train2)
print('='*40 + 'TEST SUITE 3' + '='*40)    
leaf3=tune_d_tree2(X_train3, y_train3)
```

```python

'''
max_d1, max_d2, max_d3 variables contain the maximum depth
leaf1, leaf2, leaf3 variables have min_samples_leaf values for each suite

'''
d_tree=DecisionTreeClassifier(max_depth=max_d1,min_samples_leaf=leaf1) #values specific for the training set obtained
print('='*35+' SUITE 1 (50:50) DESCISION TREE CLASSIFIER ' +'='*35)
predictive_model_1=d_tree.fit(X_train1,y_train1)
y_pred1=predictive_model_1.predict(X_test1)
cm4=confusion_matrix(y_test1, y_pred1) #stored so that it can be visualized using seaborn heatmaps
print(f'Confusion Matrix: \n{cm4}')
ac4=accuracy_score(y_test1, y_pred1)
print(f'Accuracy Score: {ac4}')
cr=classification_report(y_test1, y_pred1)
print(f'Classification Report: \n\n{cr}\n\n')
crd_tree_1=classification_report(y_test1, y_pred1, output_dict=True)
print('FEATURE IMPORTANCES:\n')
print(d_tree.feature_importances_)



d_tree=DecisionTreeClassifier(max_depth=max_d2,min_samples_leaf=leaf2) #values set for the model and traning set
print('='*35+' SUITE 2 (60:40) DESCISION TREE CLASSIFIER ' +'='*35)
predictive_model_2=d_tree.fit(X_train2,y_train2)
y_pred2=predictive_model_2.predict(X_test2)
cm5=confusion_matrix(y_test2, y_pred2)#stored so that it can be visualized using seaborn heatmaps
print(f'Confusion Matrix: \n{cm5}') 
ac5=accuracy_score(y_test2, y_pred2)
print(f'Accuracy Score: {ac5}')
cr=classification_report(y_test2, y_pred2)
print(f'Classification Report: \n\n{cr}\n\n')
crd_tree_2=classification_report(y_test2, y_pred2, output_dict=True)
print('FEATURE IMPORTANCES:\n')
print(d_tree.feature_importances_)



d_tree=DecisionTreeClassifier(max_depth=max_d3,min_samples_leaf=leaf3)
print('='*35+' SUITE 3 (80:20) DESCISION TREE CLASSIFIER ' +'='*35)
predictive_model_3=d_tree.fit(X_train3,y_train3)
y_pred3=predictive_model_3.predict(X_test3)
cm6=confusion_matrix(y_test3, y_pred3) #stored so that it can be visualized using seaborn heatmaps
print(f'Confusion Matrix: \n{cm6}')
ac6=accuracy_score(y_test3, y_pred3)
print(f'Accuracy Score: {ac6}')
cr=classification_report(y_test3, y_pred3)
print(f'Classification Report: \n\n{cr}\n\n')
crd_tree_3=classification_report(y_test3, y_pred3, output_dict=True)
print('FEATURE IMPORTANCES:\n')
print(d_tree.feature_importances_)


```
